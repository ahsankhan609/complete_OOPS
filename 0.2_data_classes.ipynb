```python
@dataclass
class Car:
    brand: str
    model: str
    production_year: int
    color: str
    price: int
```

### Exaplaination of Data Classes:

when we define the type(int) of our variables or fields(x,y), then @dataclass by default populate 3 common methods that we typically need

__init__ - It helps to initiate our class. and it takes our 2 variables, that we define in the same order

__repr__ - it is used to print out or view content of the object. It will follow this format to print results in terminal
 -> Classname(var1=value, var2=value)

__eq__ - it is used to compare two objects(instances) of the same class. it will return true if the values of the variables are same, otherwise it will return false. and it will ignore the memory location of the objects, and it will compare only the values of the variables.


```python
@dataclass
class Point:
    x: int
    y: int
```

```python
p1 = Point(1, 2)
p2 = Point(3, 4)

print(p1, p2)
print(p1 == p2)
```

#### adding default values to the variables:

```python
from dataclasses import dataclass


@dataclass
class Position:
    name: str
    lon: float = 0.0
    lat: float = 0.0
```

```python
Position('Null Island')
Position(name='Null Island', lon=0.0, lat=0.0)

Position('Greenwich', lat=51.8)
Position(name='Greenwich', lon=0.0, lat=51.8)

Position('Vancouver', -123.1, 49.3)
Position(name='Vancouver', lon=-123.1, lat=49.3)
```

`default_factory`, which gives a way to provide more complicated default values.

In fact, adding some kind of `type hint` is `mandatory` when `defining the fields` in your `data class`. Without a type hint, the field will not be a part of the data class. However, if you do not want to add explicit types to your data class, use `typing.Any`:

```python
from dataclasses import dataclass
from typing import Any


@dataclass
class WithoutExplicitTypes:
    name: Any
    value: Any = 42
```

### Adding methods to data classes:

You already know that a data class is just a regular class. That means that you can freely add your own methods to a data class. As an example, let us calculate the distance between one position and another, along the Earth’s surface.

```python
from dataclasses import dataclass
from math import asin, cos, radians, sin, sqrt


@dataclass
class Position:
    name: str
    lon: float = 0.0
    lat: float = 0.0

    def distance_to(self, other):
        r = 6371  # Earth radius in kilometers
        lam_1, lam_2 = radians(self.lon), radians(other.lon)
        phi_1, phi_2 = radians(self.lat), radians(other.lat)
        h = (sin((phi_2 - phi_1) / 2)**2
             + cos(phi_1) * cos(phi_2) * sin((lam_2 - lam_1) / 2)**2)
        return 2 * r * asin(sqrt(h))
```

```python
oslo = Position('Oslo', 10.8, 59.9)
vancouver = Position('Vancouver', -123.1, 49.3)
oslo.distance_to(vancouver)
# 7181.7841229421165
```

### use of `field()` function in data classes:

```python
# https://www.youtube.com/shorts/Zd-u433xy0c
from dataclasses import dataclass, field


@dataclass
class Name:
    first: str
    last: str
    full:  str = field(init=False)

    def __post_init__(self):
        self.full = f"{self.first} {self.last}"
```

```python
# person1 = Name(first="john", last="smith")
# print(person1.full)
person1 = Name("john", "smith")
print(person1)
print(person1.full)
```

```python
import inspect
from dataclasses import dataclass, astuple, asdict
from pprint import pprint


@dataclass(frozen=True, order=True)  # frozen will provide hash function
class Comment:
    id: int
    text: str = field(default="")
    replies:  list[int] = field(default_factory=list)


def main():
    comment = Comment(1, "I am subscriber...")
    print(comment)
    print(astuple(comment))  # output as tuple
    print(asdict(comment))  # output as dictionary
    print("\n")
    # it will give info if the all functions that are run(covered) by the data class
    pprint(inspect.getmembers(Comment, inspect.isfunction))


if __name__ == '__main__':
    main()
```

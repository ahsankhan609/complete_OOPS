### What is Object-Oriented Programming(OOP)?

* Object-oriented programming is an approach of writing programs by creating classes and objects.
* More focus is on `data` rather than `logic`.
* Its Way of organizing programs.

Python supports multiple paradigms. These are as follows :-

1) Procedural oriented paradigm
2) Functional oriented paradigm
3) Object-oriented paradigm

### Why OOP?

* To solve real-world problems effectively.
* OOP provides some features :-
1) Inheritance :- `reusability`
2) encapsulation :- `data security`
3) abstraction :- `Data hiding`

### What is Class in OOP?

* Class is a template/blueprint/prototype for creating objects.
* Every object belongs to some class.
* class is a collection of attributes and behaviours.
* technically, class is a user-defined data type.
* Email class :- email1 + email2 + email3 +email4

It tells that whenever we create an email object, it should have some attributes and behaviours. It is a template for creating objects. So, email instance attributes and behaviours will be different from ueage by usage.

````python
# some builtint data types of Python classes are:

l1: list[int] = [1, 2, 3]
s1: str = "hello world"
t1 = (1, 2, 3)


def func() -> None:
    pass


print(type(l1), type(s1), type(t1), type(func))
````

### What is an Object in OOP?

* An object in OOP represents real-life objects. ex. Email, man, student, employee etc
* Every object has two properties.
  
1) Attributes (means Data) -> heading,subject,name, recipients list of an Email Object
2) Behaviours (means Actions or Functions) -> sending,Adding attachments, deleting etc of an Email Object

### Constructor & types of Constructor

* Special method which is used for initializing objects with attributes
* Initialize default values of the object, at the time of creation.
* it is __init__ method
* First argument is `self`

There are 3 types of constructors in Python :-

* Parameterized constructor
* Non-Parameterized constructor
* Default Constructor

#### example of non-parameterized constructor

````python
class Employee:
    def __init__(self) -> None:
        self.age = 21
        self.salary = 10000
````

````python
e1: Employee = Employee()  # example of non-parameterized constructor
e2: Employee = Employee()  # example of non-parameterized constructor

print(e1.__dict__)
print(e2.__dict__)
````

#### example of parameterized constructor

````python
class Employee:
    def __init__(self, age: int, salary: int) -> None:
        self.age: int = age
        self.salary: int = salary
````

````python
e1: Employee = Employee(26, 25000)  # example of parameterized constructor
e2: Employee = Employee(24, 35000)  # example of parameterized constructor

print(e1.__dict__)
print(e2.__dict__)
````
